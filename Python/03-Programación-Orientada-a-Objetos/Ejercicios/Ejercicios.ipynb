```python
# Requerimientos funcionales:
#
# -> Retorna el resultado del cálculo solicitado sobre dos números
#
# -> params: posición 0: un número
#            posición 1: un número
#            posición 2: un texto con la operación a realizar
#                        (sum, res, div, mul)
#
# Si no se indica operación a realizar, por defecto se realiza una suma 
#
# Si alguno de los args(params) no son válidos, indicaremos que la operación
# no se puede realizar con el mensaje de Operativa no válida

def sum(a, b):
    return a+b
def res(a, b):
    return a-b
def div(a, b):
    return a/b
def mul(a, b):
    return a*b

try:
    a = int(input(f"Introduce el primer número: "))
    b = int(input(f"Introduce el segundo número: "))
    calc = input(f"Introduce la operación a realizar (suma, resta, división, multiplicación): ")

    if calc == "suma":
        print(f"{a} + {b} =  {sum(a,b)}")
        
    elif calc == "resta":
        print(f"{a} - {b} =  {res(a,b)}")

    elif calc == "multiplicación":
        print(f"{a} * {b} =  {mul(a,b)}")
        
    elif calc == "división":
        print(f"{a} / {b} =  {div(a,b)}")
        
    else:
        print(f"{a} + {b} = {sum(a,b)}")
        
except Exception as err:
    print(f"Error: Operativa no válida: {err}")
```

```python
# Ejercicio de antes versión Alejandro

operaciones = {"sum":"+", "res":"-", "div":"/", "mul":"*"}

def Calculo(n1, n2, op="sum"): 
    if(op == "sum"):
        return n1 + n2

    elif(op == "res"):
        return n1 - n2

    elif(op == "div"):
        return n1 / n2

    elif(op == "mul"):
        return n1 * n2

#################################################################################

try:   

    while True:
        numero1 = input("Introduzca el primer número: ")
        if(numero1.isdigit()):
            break

    while True:
        numero2 = input("Introduzca el segundo número: ")
        if(numero2.isdigit()):
            break

    while True:
        op = input("Introduzca la operación que desee realizar ['sum'/'res'/'div'/'mul'] (Si no introduce nada se considerará que se quiere sumar)").lower()
        if(op == "sum" or op == "res" or op == "div" or op == "mul" or op == ""):
            break


    if(op == ""):
        print(f"Cálculo de {numero1} + {numero2} = {Calculo(int(numero1),int(numero2)):1.2f}")

    else:
        print(f"Cálculo de {numero1} {operaciones[op]} {numero2} = {Calculo(int(numero1),int(numero2), op):1.2f}")    


except Exception as err:
    print(f"Operativa no válida: {err}")
```

```python
#Ejercicio 2A
#Escribir funciones para tranformar grados Celsius en Fahrenheit y viceversa.

#Formulas:
#Celsius = (5 ÷ 9) x (Fahrenheit - 32)
#Fahrenheit = (Celsius x (9 ÷ 5)) + 32

#Entrada de datos:
#Grados Celsius/Fahrenheit

#Salida de datos:
#Grados transformados

def CtoF(a):
    return (a * (9 / 5)) + 32
def FtoC(b):
    return (5 / 9) * (b - 32)

try:
    while(True):
        tipo = input(f"Indica el tipo de grados: Celsius (C) o Fahrenheit (F)").lower()
        if tipo == "c" or tipo == "celsius":
            grados = float(input(f"Indica el número de grados: "))
            print(f"Los grados en Fahrenheit son: {CtoF(grados)} F")
            break
        
        elif tipo == "f" or tipo == "fahrenheit":
            grados = float(input(f"Indica el número de grados: "))
            print(f"Los grados en Celsius son: {FtoC(grados)} C")
            break
            
        else:
            print(f"El tipo de grado no es correcto. Por favor, indique solo C (Celsius) o F (Fahrenheit).")
            break
        
except:
    print(f"El número de grados no es correcto.")
```

```python
#Ejercicio 2A
#Escribir funciones para tranformar grados Celsius en Fahrenheit y viceversa.

#Formulas:
#Celsius = (5 ÷ 9) x (Fahrenheit - 32)
#Fahrenheit = (Celsius x (9 ÷ 5)) + 32

#Entrada de datos:
#Grados Celsius/Fahrenheit

#Salida de datos:
#Grados transformados

def grado(a,b):
    if b == "c" or b == "celsius":
        print(f"Los grados en Fahrenheit son: {(a * (9 / 5)) + 32} F")
    elif b == "f" or b == "fahrenheit":
        print (f"Los grados en Celsius son: {(5 / 9) * (a - 32)} C")      

try:
    while(True):
        b=input(f"Indica el tipo de grados: Celsius (C) o Fahrenheit (F) ").lower()
        if b != "c" and b != "f" and b != "celsius" and b != "fahrenheit":
            print(f"El tipo de grado no es correcto. Tipos admitidos: Celsius (C) / Fahrenheit (F).")
            break
        a=float(input(f"Indica el número de grados: "))
        
        grado(a,b)
        break
    
except:
    print(f"El número de grados no es correcto.")
```
