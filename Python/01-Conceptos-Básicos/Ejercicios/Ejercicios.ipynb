```python
# Encuentas versión 1
# Preguntar Edad (validar)
# Preguntar puntuación de 0 a 10 (validar)
# Preguntar si es el último encuestado
# Con el último encuestado mostra el número de encuentas y la puntuación media

lista_edad=[]
lista_opiniones=[]

loop = True
while loop == True:
    
    edad = int(input("Indica tu edad: "))
    if edad <=120:
        lista_edad.append(edad)
        
        op = int(input("Indica tu opinión del curso dando un valor entre 0 y 10: "))
        if (op >= 0 and op <=10):
            lista_opiniones.append(op)
            
            ser_ultimo = input("Eres el ultimo encuestado?: y/n")
            if ser_ultimo == "y":
                print(f"El número total de encuestas es: {len(lista_edad)}")
                print(f"La puntuación media de las encuestas es: {sum(lista_opiniones)/len(lista_opiniones)}")
                loop = False
                
            else:
                print(f"Por favor, responde únicamente sí (y) o no (n)")
        
        else:
            print(f"Tu opinión debe estar entre 0 y 10")
    
    else:
        print(f"Tu edad debe estar comprendida entre 0 y 120 años")
```

```python
# Hucha version 1
# Preguntamos por la acción: Insertar dinero // Romper la hucha
# Si rompemos la hucha, mostramos el importe total
# Si insertamos: 50 // 100 // 200 // 500 // 1000
# Después de insertar, mostramos el importe total y el número de monedas de cada tipo

hucha = {"50":0, "100":0, "200":0, "500":0, "1000":0}

loop = True
while loop == True:
    acción = input("Decide tu acción: insertar dinero / romper la hucha:" )
    if acción == "insertar dinero":
        moneda = input(f"Indica cuánto quieres insertar: 50, 100, 200, 500, 1000: ")
            
        #if (moneda in hucha.keys()):
            #hucha[moneda] += 1
        if moneda == "50":
            hucha["50"] = hucha["50"]+1  
        elif moneda == "100":
            hucha["100"] = hucha["100"]+1  
        elif moneda == "200":
            hucha["200"] = hucha["200"]+1  
        elif moneda == "500":
            hucha["500"] = hucha["500"]+1  
        elif moneda == "1000":
            hucha["1000"] = hucha["1000"]+1
        else:
            print(f"La moneda {moneda} no es de un valor valido.")
        
        for key in hucha.keys():
            print(f"La hucha tiene {hucha[key]} monedas de {key}") # forma mas facil de indicar el numero de monedas que la de abajo
            totall = int(key) * hucha[key]
            print(f"Total de monedas: {totall}")
    
    elif acción == "romper la hucha":
        print(f"El número de monedas de 50 es: {hucha["50"]}")
        print(f"El número de monedas de 100 es: {hucha["100"]}")
        print(f"El número de monedas de 200 es: {hucha["200"]}")
        print(f"El número de monedas de 500 es: {hucha["500"]}")
        print(f"El número de monedas de 1000 es: {hucha["1000"]}")
        total = (hucha["50"]*50) + (hucha["100"]*100) + (hucha["200"]*200) + (hucha["500"]*500) + (hucha["1000"]*1000) 
        print(f"El importe total es: {total} euros.")
        break
        
    
```

```python
# hucha versión de borja

from os import system

hucha = {"50": 0, "100": 0, "200": 0, "500": 0, "1000": 0}

total = 0

try:
    while (True):        
        accion = input("R para romper hucha, I para insertar monedas: ").upper()

        if (accion == "I"):
            while (True):

                moneda = input("Valor de la moneda (50, 100, 200, 500 o 1000): ")
                if (moneda in hucha.keys()):
                    system("cls")
                    hucha[moneda] += 1
                    total = 0
                    
                    for key in hucha:
                        print(f"{hucha[key]} monedas de {key}")
                        total += int(key) * hucha[key]

                    print(f"\nImporte total: {total:1.2f}\n")
                    break

                else:
                    print(f"La moneda {moneda} no es de un valor valido.")

        elif(accion == "R"):
            system("cls")
            print(f"\nImporte total: {total:1.2f}\n")
            break

except Exception as err:

    print("No se puede mostrar información sobre la hucha.")
    print(err)
```

```python
#######################################################################
# Escribe un programa que sume los números pares
#
# Requerimientos:
#
#   -> Pregunta 10 veces un número al usuario
#   -> Al finalizar muestra la suma de los números pares
#
#######################################################################

nums = []
try:
    for n in range(10):
        n = float(input("Introduce un número: "))
        if n%2 == 0:
            nums.append(n)
        else:
            continue
        
    print(f"La suma de los números pares es: {sum(nums):1.0f}")
except:
    print (f"El número introducido no es válido. Por favor, introduce solo números.")
```

```python
#######################################################################
# Escribe un programa que cuente vocales
#
# Requerimientos:
#
#   -> Pregunta una frase al usuario
#   -> Cuenta el número de cada vocal tanto en mayúsculas como en
#      minúsculas
#
#######################################################################

vocales = {"a":0, "e":0, "i":0, "o":0, "u":0}

frase = input("Escribe una frase: ").lower()
for letra in frase:
    if letra in vocales:
        vocales[letra] += 1
print(vocales)
```

```python
#######################################################################
# Escribe un programa que adivine un número
#
# Requerimientos:
#
#   -> Se calcula un número de forma aleatoria entre 1 y 100
#   -> Pregunta el número hasta que el usuario lo adivine
#   -> Cuando el usuario no acierta hay que mostrar un mensaje
#      más o menos y cuando este a menos de 10 número hay que
#      el mensaje caliente caliente
#
#######################################################################

import random

# Generar un número aleatorio entre 1 y 100
numero_aleatorio = random.randint(1, 100)

print (f"Se ha generado un número aleatorio entre 1 y 100. Trata de adivinarlo.")
while(True):
    numero = input("Introduce un número entero: ")
    if numero.isdigit()==True:
        n=int(numero)
        if n == numero_aleatorio:
            print(f"Acertaste! El número era {numero_aleatorio}.")
            break
        elif n in range(numero_aleatorio-10, numero_aleatorio+10):
            print(f"Caliente caliente!")
        elif n>numero_aleatorio:
            print(f"Menos.")
        elif n<numero_aleatorio:
            print(f"Más.")    
    else:
        print (f"Por favor, introduce solo números enteros.")
        break
```

```python
#######################################################################
# Escribe un programa que determine si una frase es un palíndromo
#
# Requerimientos:
#
#   -> Pregunta una frase al usuario
#   -> Responde: La frase SI/NO es un palíndromo
#   
# Los PALÍNDROMOS son palabras o frases que se leen igual de izquierda
# a derecha, que de derecha a izquierda. Por ejemplo:
#
#   Allí ves Sevilla
#   A mí me mima
#
#######################################################################


frase_completa = input("Escribe tu frase: ").lower()
frase_no_esp = frase_completa.replace(" ", "")

if frase_no_esp == frase_no_esp[::-1]:
    print(f"La frase SÍ es un palíndromo")
else:
    print(f"La frase NO es un palíndromo")

# faltan los replace con los ACENTOS
```
